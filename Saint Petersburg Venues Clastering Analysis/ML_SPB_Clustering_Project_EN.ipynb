## Introduction/Business Problem section

Prior setting for this project is hypothetical restaurateur’s request for market analysis. The well-known restaurateur decided to open a new venue in Saint Petersburg, Russia. It would be obvious suggestion that the most common place for it is the center of the city. Area is one of the most major aspect of the site selection as Russian cities are widely spaced along the land. Also, there are too many competitive restaurants may be placed there, so it is important to find the best possible place to get more profit for effective pay back.

The problem of this project is to define better district for opening a new high cuisine restaurant based on points:
1.	Focus on customers with high-income
2.	Total population of districts
3.	Less possible competitive venues along the area

```python
def HTMLTableToList(table, 
                   toErase = []
                  ):
    '''
    This funtion transform BeautifulSoup tag object with table
    into list of columns. This version removes references from
    column names only.
    TODO: Add references check
    
    :param bs4.tag table: target tag.
    :param list toErase: list of symbls which should be removed 
                         from the results (Default value = [])
    
    :return: list (Content of the table), 
             list (Content of the column row)
    '''
    
    # Lists to store table's data
    ColNames = []
    Content = []

    # Prepare data
    rows = table.find_all('tr') # parse html as rows with tag 'tr'
    rawColNames = rows[0].find_all('th') # parse 1st row as tagged list

    # Appending list with names of columns
    for i in rawColNames:
    # removing references from the table row
        if i.find('sup') is not None:
            i.sup.decompose()
        ColNames.append(re.sub("<.*?>", "", str(i)))

    # Create a list of rows:
    # i - for each row, 
    # j - for splitted elements from i-element of rows
    for i in rows:
        if i.td is not None:
            tempRow = []
            for j in range(0, len(i.find_all('td'))):
                rawRow = re.sub("<.*?>", "", str(i.find_all('td')[j]))
                # Clean row value from formating marks
                for k in toErase:
                    rawRow = rawRow.replace(k, "")
                tempRow.append(rawRow)
            Content.append(tempRow)
    
    return(Content, ColNames)
```

To solve this problem let's try 3 different approach:
1. Creating Voronoi diagram for searching vertices between districts
2. Use analytical geometry function to determine best possible radius
3. Adjust current values

We'll try example from: https://forum.sources.ru/index.php?showtopic=398331

 (metric - sqrt{(x-x[i])^2+(y-y[i])^2}-R[i], where R[i] -- distance to the nearest center)

### Creating Voronoi diagram for district space segmentation

```python
# Convert lat/lng coordinates for the plot
points = []
for i in range(0, len(spb_points.index)):
    points.append([spb_points.iloc[i][2], spb_points.iloc[i][1]])

# Building plot for marking districts
vor = Voronoi(points)
fig = voronoi_plot_2d(vor)
#fig = voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=3, line_alpha=0.6, point_size=2)
plt.show()

# Verticals as be
vertices = vor.vertices.tolist()
```

```python
# Finding nearest Verticals point with geodesic
n_vert_coord = []
    
for i in range(0, len(spb_points.index)):
    closest_point = [0,0,0]
    dot_1 = (spb_points.iloc[i][1], spb_points.iloc[i][2])
    for j in range(0, len(vertices)):
        dot_2 = (vertices[j][1], vertices[j][0])
        dist = geodesic(dot_1, dot_2).m
        if dist > 0 and ((closest_point[2] == 0) or (closest_point[2] >= dist)):
            closest_point = [vertices[j][1], vertices[j][0], dist]
    n_vert_coord.append(closest_point)
n_vert_coord

closest_vert_df = pd.DataFrame(n_vert_coord, columns=['ClosestVertLat', 'ClosestVertLng', 'VertDist'])
spb_points = spb_points.merge(closest_vert_df, left_index=True, right_index=True)
spb_points.head()
```

### Calculating nearest District for calculating better radius value

```python
# Finding nearest districts with geodesic
n_dist_coord = []
    
for i in range(0, len(spb_points.index)):
    closest_point = [0,0,0,0]
    dot_1 = (spb_points.iloc[i][1], spb_points.iloc[i][2])
    for j in range(0, len(spb_points.index)):
        dot_2 = (spb_points.iloc[j][1], spb_points.iloc[j][2])
        dist = geodesic(dot_1, dot_2).m
        if dist > 0 and ((closest_point[2] == 0) or (closest_point[2] >= dist)):
            closest_point = [spb_points.iloc[j][1],
                             spb_points.iloc[j][2], 
                             dist, 
                             int(spb_points.iloc[j][0])]
    n_dist_coord.append(closest_point)

closest_df = pd.DataFrame(n_dist_coord, columns=['ClosestDstLat', 'ClosestDstLng', 'Dist', 'ClosestDNumber'])
spb_points = spb_points.merge(closest_df, left_index=True, right_index=True)
```

Now we have dataframe with data of coordinates and distance between each District center and:
1. Closest District
2. Closest vertice of Voronoi diagram

```python
# Calculate best radius for each distict
radius_func = []
for i in range(0, len(spb_points)):
    x = geodesic((spb_points.iloc[i][1], 0),
                 (spb_points.iloc[i][6], 0)).m
    y = geodesic((0, spb_points.iloc[i][2]),
                 (0, spb_points.iloc[i][7])).m
    dist = spb_points.iloc[i][8]
    rad = math.sqrt(x ** 2 + y ** 2) - dist
    radius_func.append(rad)

radius_func[:5]
```

```python
r_values = pd.Series(radius_func)
copy_spb_df = spb_df.copy()

try:
    copy_spb_df.drop(axis=1, columns='Radius_f', inplace=True)
except:
    print('No Radius_f. Resuming...')
try:
    copy_spb_df.drop(axis=1, columns='Radius_v', inplace=True)
except:
    print('No Radius_v. Resuming...')

copy_spb_df.insert(loc=7, column='Radius_f', value=r_values)
copy_spb_df.insert(loc=8, column='Radius_v', value=spb_points.VertDist)
```

### Let's find out best K-param for KMeans clustering process

```python
# set number of clusters

scores = []
for i in range(2, 10):
    kclusters = i
    spb_clustering_df = X
    # run k-means clustering
    kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(X)
    # check cluster labels generated for each row in the dataframe
    db_score = davies_bouldin_score(X, kmeans.labels_)
    ch_score = metrics.calinski_harabasz_score(X, kmeans.labels_)
    s_score = metrics.silhouette_score(X, kmeans.labels_, metric='euclidean')
    scores.append([i, db_score, ch_score, s_score])
scores
```

#### Hint!
1. db_score - lower means better cluster separation

2. ch_score - higher means better cluster definition

3. higher Silhouette Coefficient score relates to a model with better defined clusters
